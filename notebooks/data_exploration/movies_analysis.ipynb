This cell loads the main libraries needed for working with data (Pandas) and for plotting graphs (Seaborn and Matplotlib).



```python
# Import required libraries for data manipulation and visualization
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

This cell runs a SQL query to extract the movie names and their ratings from the movies table, then loads the result into a Pandas DataFrame.

```python
from db_connection import get_dataframe

# SQL query to retrieve movie ratings
query = """
SELECT name, rating
FROM movies
"""

# Execute the query and store result in a DataFrame
movies_df = get_dataframe(query)
movies_df.head()
```

This cell helps identify if there are missing or inconsistent rating values before proceeding with the visualization.

```python
# Check for null or invalid ratings
print("Missing values:\n", movies_df.isnull().sum())
print("\nUnique ratings:", movies_df["rating"].unique())
```

This cell creates a histogram to visualize how the movie ratings are distributed. The KDE curve shows the probability density.

```python
# Set the visual style
sns.set(style="whitegrid")

# Plot a histogram of ratings
plt.figure(figsize=(10, 6))
sns.histplot(movies_df["rating"], bins=20, kde=True, color="skyblue")
plt.title("Distribution of Movie Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Movies")
plt.show()
```
